```python
import shutil
import os


def copy_directory(src, dst):
    # Check if destination directory exists
    if not os.path.exists(dst):
        os.makedirs(dst)
    # Copy each file/subdirectory
    for item in os.listdir(src):
        s = os.path.join(src, item)
        d = os.path.join(dst, item)
        # If item is a directory, recurse
        if os.path.isdir(s):
            copy_directory(s, d)
        else:
            shutil.copy2(s, d)  # can preserve file data


# Usage
src_dir = "/content/drive/Shareddrives/extradrive/Non_recycle_dataset"
dst_dir = "/content/drive/Shareddrives/extradrive/RecycleDataset_copied1"
copy_directory(src_dir, dst_dir)
```
